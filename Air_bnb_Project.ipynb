# CLEAN THE DATA

*Steps to be Taken for Data Cleaning*

* Drop duplicated records.
* Drop house_rules and license columns with insufficient data.
* Remove all dollar signs in the price and service fee columns.
* Remove all commas from the price and service fee columns.
* Rename the price and service fee columns to include a dollar signs.
* Drop all records with missing values.
* Change all mismatched data types to the appropriate once.
* Get rid of outliers in the 'availability 365' column data.


```python
df['last review']=pd.to_datetime(df['last review'], errors='coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month': 0,'last review':df['last review'].min()}, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['NAME','host name'],inplace=True)
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=['license','house_rules'],errors='ignore')
```

```python
df.isnull().sum()
```

```python
# Remove dollar sign from price and fee
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

Q.1. How listings were distibuted?

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=20, kde=True,color='red')
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

The histogram shows a fairly even distribution of listing prices across different price ranges, indicating no particular concentration of listings in any specific price range. The KDE line helps visualize this even spread more clearly, confirming that the dataset contains listings with a wide variety of prices.

Q.2. How are different types of rooms are distributed?

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='room type', data=df, palette='viridis',color='hotpink')
plt.title('Distribution of Room Types')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

From above count plot we can easily figure out that Private room and Entirehome/aparment is too in number as compared to Shared room or Hotel rooms combined.Hosts are mainly foucuses on Private room or Entire home to Encounter Privacy.

Q.3. How are listings distributed across different neighborhoods?

```python
plt.figure(figsize=(12, 8))
sns.countplot(y='neighbourhood group',data=df,color='cyan',order=df['neighbourhood group'].value_counts().index)
plt.title('Distribution of Listings Across Neighborhoods')
plt.xlabel('Neighborhood')
plt.ylabel('Count')
```

Q.4. What is the relation between price and room type?

```python
from numpy._core.defchararray import title
plt.figure(figsize=(10, 6))
sns.boxplot(x='room type', y='price', hue='room type',data=df, palette='viridis')
plt.title('Price Vs  Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.legend(title='Room Type')
plt.show()
```

From above graph we can conclude that almost all type of rooms and apartments are having same price range that is 350 to 900.

Q.5. How the number of reviews change over time ?


```python
df['last review']=pd.to_datetime(df['last review'])
review_over_time=df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12, 6))
review_over_time.plot(kind='line', color='green')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```

As time goes reviews are droping down significantly bacause their is too number of review in 2019, but after it, number of review droped down to almost zero.So we can conclude that after Covid-19 pandemmic people are not aware of reviewing.

Q.6. Is there a relationship between the construction year of property and price?

```python
mean_price = df.groupby('Construction year')['price'].mean()

mean_price.plot(kind='line')
plt.xlabel('Construction Year')
plt.ylabel('Average Price')
plt.title('Average Price for Properties in each Construction Year')
plt.show()
```

Q.7. Who are the top 10 hosts by calculated host listing count?

```python
hosts = df.groupby('NAME')['calculated host listings count'].sum().sort_values(ascending = False).nlargest(10).to_frame()
hosts_bar = plt.bar(hosts.index, hosts.loc[:, 'calculated host listings count' ]);
plt.bar_label(hosts_bar, label = hosts.loc[:, 'calculated host listings count'], label_type = 'edge', padding = 3);
plt.xlabel('Host Name');
plt.ylabel('Calculated Host Listings Count');
plt.title('Top 10 Hosts by Calculated host Listings Count');
```
